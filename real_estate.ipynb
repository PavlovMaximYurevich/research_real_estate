```python
def which_floor(row):
    floor = row['floor']
    floors_total = row['floors_total']
    
    
    if floor == 1:
        return "First floor"
    if floors_total - floor == 0:
        return "Last floor"
    return "Other floor"
```

```python
data['which_floor'] = data.apply(which_floor, axis=1)
```
